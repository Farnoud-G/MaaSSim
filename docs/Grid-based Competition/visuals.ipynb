```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set global seaborn and matplotlib parameters
sns.set(rc={"figure.dpi": 200, 'savefig.dpi': 200})
sns.set_palette('tab10')
sns.set(font_scale=0.5)


# Define the range of values for x and y axes
x = np.array([1, 2, 3])
y = np.array([1, 2, 3])

values = np.array([[1, 2, 3],
                   [2, 3, 4],
                   [3, 4, 5]])

# Create the meshgrid
X, Y = np.meshgrid(x, y)

# Visualize the meshgrid
plt.figure(figsize=(3, 3))  # Adjust figure size as per your requirement
plt.scatter(X, Y, c=values, cmap='bwr', s=60)
plt.title("Title")
plt.xlabel('P1 fare [€/km]')
plt.ylabel('P2 fare [€/km]')

# Customize ticks and grid lines
plt.xticks(x)  # Only show ticks at [1, 2, 3] on the x-axis
plt.yticks(y)  # Only show ticks at [1, 2, 3] on the y-axis
plt.grid(True, color='black', which='both', linestyle='-', linewidth=0.5, )  # Show grid lines for both major and minor ticks

e = 0.08
plt.xlim(x.min()-e, x.max()+e)
plt.ylim(y.min()-e, x.max()+e)

cbar = plt.colorbar()
cbar.set_label('Values')

# plt.gca().set_facecolor('white')
plt.tight_layout()  # Ensures everything fits within the figure area properly
plt.show()
```
