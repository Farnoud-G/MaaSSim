```python
# remember that the attitude is always in 0<attitude<1 range

pre_att_i = 0.5 # attitude toward alternative i
pre_att_j = 1  # attitude toward alternative j
B_learning = 1  #learning speed in range 0<B_learning<1
mu = 10
att_alt_choice = 0.5

x = []
y = []
p = []

for t in range(1,12):
    
    y.append(pre_att_i)
        
    pre_cu_i = ln((1/pre_att_i)-1)  # first we calculate the previous cumulative utility from previous attitude
    cu_i = pre_cu_i + pre_att_i - pre_att_j # we update the cumulative utility
    att_i = 1/(1+math.exp(B_learning*cu_i))  # we calculate the new attitude wrt updated cumulative utility
    prob = (math.exp(mu*att_i))/(math.exp(mu*att_i) + math.exp(mu*att_alt_choice))
    
    x.append(t)
    p.append(prob)
    
    pre_att_i = att_i

fig, ax = plt.subplots(figsize=(8,8))
sns.lineplot(x=x, y=y)
```

```python
# Learning from experiences
fig, ax = plt.subplots(figsize=(16,5))

RW = 40
E_u = 0.01
M_u = 0.5
W_u = 0.5
# U = 0
B_learning = 1
a = 1
x = []
ye = []; ym = []; yw = []; yu = []

for t in range(0,50):
    
    x.append(t)
    ye.append(E_u)
    ym.append(M_u)
    yw.append(W_u)
    
    if t<20:
        M_u = 1/(1+math.exp(B_learning*(ln((1/M_u)-1) + 1.5*(M_u - 1))))
    if 9<t<30:
        W_u = 1/(1+math.exp(B_learning*(ln((1/W_u)-1) + a*(W_u - 1))))
    if 29<t<50:
        W_u = 1/(1+math.exp(B_learning*(ln((1/W_u)-1) + a*(W_u - 0))))
    if 19<t<40:
        E_u = 1/(1+math.exp(B_learning*(ln((1/E_u)-1) + 1*(RW - 60)/abs(RW))))
    if 39<t<50:
        E_u = 1/(1+math.exp(B_learning*(ln((1/E_u)-1) + 1*(RW - 20)/abs(RW))))
    
    U = (1/3)*E_u + (1/3)*M_u + (1/3)*W_u
    yu.append(U)
    

sns.lineplot(x=x, y=ye, label='Experience')
sns.lineplot(x=x, y=ym, label='Marketing')
sns.lineplot(x=x, y=yw, label='WOM')
sns.lineplot(x=x, y=yu, label='Perceived utility')
```
