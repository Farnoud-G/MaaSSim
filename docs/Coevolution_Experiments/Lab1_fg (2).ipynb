```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
import random
```

### 1. Import driveable network for your city

```python
place_name = 'Piedmont, California, USA'
G = ox.graph_from_place(place_name, network_type='drive')
```

```python
area = ox.geocode_to_gdf(place_name)
area.plot()
```

```python
poly = area.geometry[0]
G = ox.graph_from_polygon(polygon = poly, network_type = 'drive')
```

```python
fig, ax = ox.plot_graph(G)
```

```python
node_ide = 53017091
G.nodes[node_ide]
```

```python
file_name = 'Delft.graphml'
ox.save_graphml(G, filepath=file_name)
```

```python
node_list = list(G.nodes)
o = random.choice(node_list)
d = random.choice(node_list)
```

```python
route = nx.shortest_path(G, o, d, weight='length')
fig, ax = ox.plot_graph_route(G, route, node_size=0)
```

```python
# find the route between these nodes then plot it
route = nx.shortest_path(G, o, d, weight='length')
fig, ax = ox.plot_graph_route(G, route, node_size=0)
print("Shorthest path from node {} to node {} with distance: {} m".format(o,d,nx.shortest_path_length(G, o, d, weight='length')))
```
