```python
ams = pd.read_csv('Amsterdam_requests_new.csv')
ams.treq = ams.apply(lambda row: pd.Timestamp(row.treq), axis=1)

s = pd.Timestamp('1900-01-01 12:00:00')
e = pd.Timestamp('1900-01-01 18:00:00')
ams = ams[ams.treq>s]
```

```python
window_number = (e.hour-s.hour)/0.5
t = s
sample_size = 5
seed = 1

for i in range(0,int(window_number)):
    t_new = t+pd.Timedelta(minutes=30)
    df = ams.loc[(ams.treq<t_new) & (ams.treq>t)]
    df = df.sample(sample_size, random_state= seed)
    df.to_csv('Olga/trips{}_{}.csv'.format(t.strftime('%H-%M-%S'),t_new.strftime('%H-%M-%S')))
    t = t_new
```
